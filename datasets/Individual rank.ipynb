```python
import pandas as pd
import urllib.request, json

url = "https://ifsc.egroupware.net/egw/ranking/json.php?person=13208"

with urllib.request.urlopen(url) as url:
    data = json.loads(url.read().decode())

pd.io.json.json_normalize(data['results'])
```

```python
results = pd.io.json.json_normalize(data['results'])
results['rank'] = results['rank'].astype(int)
results['date'] = pd.to_datetime(results['date'])
```

```python
results.plot(x='date', y='rank')
```
